```python
fig = plt.figure(figsize=(10, 10))
plt.matshow(movie_corr, fignum = fig.number)
plt.colorbar()
plt.xticks(range(movie_corr.shape[1]), movie_corr.columns, fontsize=14, rotation=45)
plt.yticks(range(movie_corr.shape[1]), movie_corr.columns, fontsize=14)
plt.show()
```
