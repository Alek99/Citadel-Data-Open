```python
pivoted = ratings.pivot(index='userId', columns='movieId', values='rating').fillna(0)
pivoted.shape
```

```python
pivoted = pivoted.loc[:, pivoted.sum(axis=0) > 100]
pivoted.shape
```

```python
pivoted
```

```python
testvector = [0,0,0, 1, 4, 1, 0, 0, 0, 0, 0, 0, 0, 3, 4, 5, 5]

distances = []
for row in np.array(pivoted):
    distances.append(distance.euclidean(testvector, row))
```

```python
min_distance_index = distances.index(min(distances))
nearestmatch = pivoted.iloc[min_distance_index]
```
